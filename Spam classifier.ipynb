```python
import pandas as pd
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None) classifier
pd.set_option('display.max_columns', None)
messages = pd.read_csv('SMSSpamCollection.txt', sep='\t',
                           names=["label", "message"])  #columns are tab seperated.////as we do not have any columns we specify the  column names as labels and messages.
```

```python
messages
```

```python
import re
import nltk
```

```python
#Data cleaning and preprocessing
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i]) #removing unessary charactrs
    review = review.lower()   #convert everything lower case
    review = review.split()   #put the result in a list
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]  #convert the words to stem except for stopwords
    review = ' '.join(review)   #put everything together
    corpus.append(review)
```

```python
print(corpus)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000) # selecting top 5000 words
X = cv.fit_transform(corpus).toarray() #
```

Independent feature
--

```python
print(X.shape)
print(X)
```

```python
#output data...we have ham/spam,, we convert it to dummy values so that machine can understand
y=pd.get_dummies(messages['label'])
print(y)  #if true then it says it is ham/spam
```

dependent feture
--

```python
#reducing it to a list
y1=y.iloc[:,1].values
print(y1)
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size = 0.20, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
```

```python
# Training model using Naive bayes classifier

from sklearn.naive_bayes import MultinomialNB     #navie bayes library
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)
```

```python
print(y_pred)
```

```python
#comparing y_pred and y1
from sklearn.metrics import confusion_matrix
condusion_m=confusion_matrix(y_test,y_pred)
condusion_m  #this tells us that 942+152 are correctly predicted
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
